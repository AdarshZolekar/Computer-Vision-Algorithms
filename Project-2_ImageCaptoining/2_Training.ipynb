### Question 1

**Question:** Describe your CNN-RNN architecture in detail.  With this architecture in mind, how did you select the values of the variables in Task 1?  If you consulted a research paper detailing a successful implementation of an image captioning model, please provide the reference. 

**Answer:** The CNN-RNN architecture consits of a CNN encoder and RNN decoder. The CNN encoder is the [50-layer Resnet](https://arxiv.org/pdf/1512.03385.pdf) model that is a good pre-trained convolutional neural network on ImageNet. The network serves as a feature encoder for images to be captioned. The last fully connected layer of encoder is dropped and replaced by a new fully connected layer transforming the final feature map to an encoding. This encoded feature map is the input for the RNN decoder. Reccurent neural network (RNN) is useful in the tasks with sequential nature of data such as machine translation or speach recognition. Usually the RNN network stores only some last states of the network (for example, three last words in the text sentence or speech). The RNN network can include the advanced and sophisticated component, the layer with long memory proposed in 1997 by [S.Hochreiter and J.Schmidhuber](https://www.mitpressjournals.org/doi/abs/10.1162/neco.1997.9.8.1735). This layer is called [Long short-term memory (LSTM)](https://colah.github.io/posts/2015-08-Understanding-LSTMs/). We first use the CNN encoder output concatenated with the "START" token as the initial input for the RNN decoder. In the RNN decoder, we prepare the tensor from incoming feature map and captions. This tensor together with hidden state is the input for the LSTM layer. We apply a fully-connected layer to derive a softmax probability. The word with the highest probability in our entire vocabulary is the the predicted word. Then, this predicted word become the input for the next prediction. We continue so until the network generated the "STOP" token, which is the end of the sentence.

( We continue so until we generated a caption of max length, or the network generated the "STOP" token, which is the end of the sentence.)

To select the hyperparameters I consulted with papers ["Show, Attend and Tell"](https://arxiv.org/pdf/1502.03044.pdf), ["Empirical Exploration of RNN"](http://proceedings.mlr.press/v37/jozefowicz15.pdf), and I also tested several models with different sets of hyperparameters. Two criteria were used to select a set of hyperparameters:
  * the least value of the "_loss_" function, 
  * visualization of several images with the predicted captions. 

Finally, the following parameters (in bold) are selected:
  * batch_size = **128** # tried also 32, 64, 128, 256,   
  * vocab_threshold = **4** # tried also 5,6,         
  * embed_size = **256** # tried also 512,           
  * lr = **0.001**, # tried also 0.001, 0.005,                        
  * criterion = **CrossEntropyLoss**, 
  * optimizer = **Adam** # tried also ASGD, Adagrad.

````python
# Watch for any changes in model.py, and re-load it automatically.
% load_ext autoreload
% autoreload 2

#sys.path.append(
import torch
print('version torch: ', torch.__version__)
import torch.nn as nn
from torchvision import transforms
import sys
#sys.path.append('/opt/cocoapi/PythonAPI')
from pycocotools.coco import COCO
from data_loader import get_loader
from model import EncoderCNN, DecoderRNN
import math
````

````python
## TODO #1: Select appropriate values for the Python variables below.
batch_size = 128 # 32 # 64 # 128 # 64 # 96 ## 16 # 32 ## 64 ## previous: 128          # batch size
vocab_threshold = 4 # 6 # 5 # 4        # minimum word count threshold
vocab_from_file = False    # if True, load existing vocab file
embed_size = 256 ## 512 ## 256 ## 512 ## prev: 256           # dimensionality of image and word embeddings
hidden_size = 512         # number of features in hidden state of the RNN decoder
clip_value = 2             # the maximum gradient value for clipping
num_epochs = 3 # 2 ##  prev: 5            # number of training epochs
save_every = 1             # determines frequency of saving model weights
print_every = 100  ## prev 100         # determines window for printing average loss
log_file = 'training_log.txt'       # name of file with saved training loss and perplexity

# (Optional) TODO #2: Amend the image transform below.
transform_train = transforms.Compose([ 
    transforms.Resize(256),                          # smaller edge of image resized to 256
    transforms.RandomCrop(224),                      # get 224x224 crop from random location
    transforms.RandomHorizontalFlip(),               # horizontally flip image with probability=0.5
    transforms.ToTensor(),                           # convert the PIL Image to a tensor
    transforms.Normalize((0.485, 0.456, 0.406),      # normalize image for pre-trained model
                         (0.229, 0.224, 0.225))])
````

````python
# Build data loader.
data_loader = get_loader(transform=transform_train,
                         mode='train',
                         batch_size=batch_size,
                         vocab_threshold=vocab_threshold,
                         vocab_from_file=vocab_from_file)
````

````python
# The size of the vocabulary.
vocab_size = len(data_loader.dataset.vocab)

# Initialize the encoder and decoder. 
encoder = EncoderCNN(embed_size)
decoder = DecoderRNN(embed_size, hidden_size, vocab_size)

# Move models to GPU if CUDA is available. 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder.to(device)
decoder.to(device)

# Define the loss function. 
criterion = nn.CrossEntropyLoss().cuda() if torch.cuda.is_available() else nn.CrossEntropyLoss()
#criterion = nn.MSELoss().cuda() if torch.cuda.is_available() else nn.MSELoss()

# TODO #3: Specify the learnable parameters of the model.
params = list(decoder.parameters()) + list(encoder.embed.parameters()) 

# TODO #4: Define the optimizer.
optimizer = torch.optim.Adam(params, lr=0.001)
#optimizer = torch.optim.Adagrad(params, lr=0.001)
#optimizer = torch.optim.Adam(params, lr=0.01)
#optimizer = torch.optim.ASGD(params, lr=0.001)
#optimizer = torch.optim.ASGD(params, lr=0.005)
#optimizer = torch.optim.ASGD(params, lr=0.01)

# Set the total number of training steps per epoch.
total_step = math.ceil(len(data_loader.dataset.caption_lengths) / data_loader.batch_sampler.batch_size)
````

````python
import torch.utils.data as data
import numpy as np
import os
import requests
import time
from workspace_utils import active_session
````

````python
# Open the training log file.
f = open(log_file, 'w')

#old_time = time.time()
    
print('total_step: ', total_step)    
for epoch in range(1, num_epochs+1):

    for i_step in range(1, total_step+1):

        #print('i_step in range total_step: ', i_step)

        # Randomly sample a caption length, and sample indices with that length.
        indices = data_loader.dataset.get_train_indices()
        # Create and assign a batch sampler to retrieve a batch with the sampled indices.
        new_sampler = data.sampler.SubsetRandomSampler(indices=indices)
        data_loader.batch_sampler.sampler = new_sampler

        # Obtain the batch.
        images, captions = next(iter(data_loader))

        # Move batch of images and captions to GPU if CUDA is available.
        images = images.to(device)
        captions = captions.to(device)

        # Zero the gradients.
        decoder.zero_grad()
        encoder.zero_grad()

        # Pass the inputs through the CNN-RNN model.
        features = encoder(images)
        outputs = decoder(features, captions)

        #print('type(captions): ', type(captions), ', type(captions.float()): ', type(captions.float()) )
        #print('type(captions.view(-1)): ', type(captions.view(-1)))
        # Calculate the batch loss.
        loss = criterion(outputs.contiguous().view(-1, vocab_size), captions.view(-1))

        # Backward pass.
        loss.backward()
        
        # Apply gradient clipping.
        torch.nn.utils.clip_grad_value_(decoder.parameters(), clip_value)
        
        # Update the parameters in the optimizer.
        optimizer.step()
        
        # Get training statistics.
        progress = ((epoch-1)*total_step + i_step) / (num_epochs*total_step) * 100
        stats = '%.2f%% Epoch [%d/%d], Step [%d/%d], Loss: %.4f, Perplexity: %5.4f' % (progress, epoch, \
                                                    num_epochs, i_step, total_step, loss.item(), np.exp(loss.item()))

        # Print training statistics (on same line).
        print('\r' + stats, end="")
        sys.stdout.flush()

        # Print training statistics to file.
        f.write(stats + '\n')
        f.flush()
        
        # Print training statistics (on different line).
        if i_step % print_every == 0:
            print('\r' + stats)
            torch.save(decoder.state_dict(), os.path.join('models_v10', 'decoder-%d.pkl' % epoch))
            torch.save(encoder.state_dict(), os.path.join('models_v10', 'encoder-%d.pkl' % epoch))

# Save the weights.
    if epoch % save_every == 0:
        torch.save(decoder.state_dict(), os.path.join('./models_v10', 'decoder-%d.pkl' % epoch))
        torch.save(encoder.state_dict(), os.path.join('./models_v10', 'encoder-%d.pkl' % epoch))

# Close the training log file.
f.close()
````
